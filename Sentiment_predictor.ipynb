Data cleaning

```python
import re
import nltk

nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

all_stopwords = stopwords.words('english')
all_stopwords.remove('not')
```

```python
corpus = []

for i in range (0,100):
  review = re.sub('[^a-zA-Z]', ' ',dataset['Review'][i])
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
  review = ' '.join(review)
  corpus.append(review)

```

Data transofrmation


```python
#loading BoW dictionary
from sklearn.feature_extraction.text import CountVectorizer
import pickle

cvFile_path = '/content/drive/MyDrive/Colab Notebooks/sentimentAnalysis/cv-transform.pkl'
cv = pickle.load(open(cvFile_path, 'rb'))
```

```python
X_new = cv.transform(corpus).toarray()
X_new.shape
```

Prediction (via sentiment classifier)


```python
import joblib
classifier = joblib.load('/content/drive/MyDrive/Colab Notebooks/sentimentAnalysis/classifier_sentimentModel')
```

```python
Y_pred = classifier.predict(X_new)
print(Y_pred)
```

```python
dataset['Predicted_label'] = Y_pred.tolist()
dataset.head()
```
